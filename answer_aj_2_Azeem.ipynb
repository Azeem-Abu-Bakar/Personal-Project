```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('applied_job_sample.csv')
```

```python
df = df[df['salary_min'] & df['salary_max']>99]
```

```python
df['salary'] = (df['salary_min'] + df['salary_max'])/2
df
```

```python
df = df.groupby(['state'])[['salary','job_id']].agg([('average_salary','mean'), ('median_salary','median'),
                                                           ('total_application','count')])
df
```

```python
df = df.drop(columns=['job_id'])
df
```
